**Data Preprocessing**

```python

def categorize_experience(years):
    if years <= 3:
        return 'Entry-level'
    elif years <= 7:
        return 'Mid-level'
    else:
        return 'Senior'

Salary['Experience Level'] = Salary['Experience Years'].apply(categorize_experience)
```

**Define Target Variabel(y) and Feature Variable(x)**

```python
Salary.columns
```

```python
y=Salary['Salary']
x=Salary[['Experience Years']]
```

**Train TEST Split**

```python
from os import XATTR_CREATE
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2529)
```

**Model Evaluation**

How to select the model
sum of error square to be minimum (least square)
Salry=intercept+slope experience +error


LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```
